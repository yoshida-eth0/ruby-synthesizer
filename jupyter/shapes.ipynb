```ruby
$LOAD_PATH << File.dirname(__FILE__) + "/../lib"
$LOAD_PATH << File.dirname(__FILE__) + "/../../ruby-audio_stream/lib"

require 'synthesizer'

include Synthesizer

xs = []
ys1 = []
ys2 = []
ys3 = []
ys4 = []
ys5 = []
ys6 = []

hz = 440.0
sample = 44100.0
delta = hz / sample
pos = ShapePos.new(1, 0)
phase = 0.0

300.times {|i|
  xs << i
  ys1 << Shape::Sine[phase]
  ys2 << Shape::Sawtooth[phase]
  ys3 << Shape::Square[phase]
  ys4 << Shape::Triangle[phase]
  ys5 << Shape::RampUp[phase]
  ys6 << Shape::RampDown[phase]
  phase = pos.next(delta, 0, 0)
}

traces = [
  {x: xs, y: ys1, name: "Sine"},
  {x: xs, y: ys2, name: "Sawtooth"},
  {x: xs, y: ys3, name: "Square"},
  {x: xs, y: ys4, name: "Triangle"},
  {x: xs, y: ys5, name: "RampUp"},
  {x: xs, y: ys6, name: "RampDown"},
]
Plotly::Plot.new(data: traces).show

nil
```
